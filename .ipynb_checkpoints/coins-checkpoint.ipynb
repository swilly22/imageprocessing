```python
%matplotlib inline

import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt

# BGR Format.
img = cv2.imread('images/me.png')

# convert image to RGB color for matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show image with matplotlib
plt.imshow(img)
```

```python
# find average per row, assuming image is already in the RGB format.
# np.average() takes in an axis argument which finds the average across that axis.
average_color_per_row = np.average(img, axis=0)

# find average across average per row
average_color = np.average(average_color_per_row, axis=0)

# convert back to uint8
average_color = np.uint8(average_color)
print(average_color)
```

```python
# create 100 x 100 pixel image with average color value
average_color_img = np.array([[average_color]*100]*100, np.uint8)

plt.imshow(average_color_img)
```

```python
# convert image to grayscale
img = cv2.imread('images/coins.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display image
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

```python
# threshold for image, with threshold 60
# if pixel at [x,y] is > threshold then use maxval, otherwise 0.
maxval = 255
threshold = 90
_, threshold_img = cv2.threshold(gray_img, threshold, maxval, cv2.THRESH_BINARY)

# show image
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

```python
# Adaptive threshold.
print("Using adaptive thresholding with mean of neighbourhood area.")

blockSize = 11
C = 4 # subtracted from the mean or weighted mean. Normally, it is positive but may be zero or negative as well.
threshold_img = cv2.adaptiveThreshold(gray_img, maxval, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize, C)

# show image
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

```python
# Adaptive threshold.
print("Using adaptive thresholding with weighted sum of neighbourhood values, weights are gaussian window.")
blockSize = 11
C = 3 # subtracted from the mean or weighted mean. Normally, it is positive but may be zero or negative as well.
threshold_img = cv2.adaptiveThreshold(gray_img, maxval, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize, C)

# show image
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

```python
print("Thresholding using color range.")

# open new Mondrian Piet painting photo
img = cv2.imread('images/coins.jpg')
# img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# threshold for hue channel in silver range
# silver_min = np.array([119, 86, 44], np.uint8)
copper_min = np.array([110, 80, 40], np.uint8)
copper_max = np.array([171, 135, 79], np.uint8)

threshold_copper_img = cv2.inRange(img, copper_min, copper_max)

threshold_copper_img = cv2.cvtColor(threshold_copper_img, cv2.COLOR_GRAY2RGB)

plt.imshow(threshold_copper_img)
```

```python
print("Using image mask to paint only coin")

# perform bitwise and on mask to obtain cut-out image that is not blue
masked = cv2.bitwise_and(img, threshold_copper_img)

plt.imshow(masked)
```

```python
def MatchRatios(ratios, hints):
    #Diameters:
    #Coin

    #		0.1		0.5		1.0		2.0		5.0		10.0

    #0.1		1		0.846	1.222	1.018	0.916	0.956

    #0.5		1.181	1		1.444	1.203	1.083	1.130

    #1.0		0.818	0.692	1		0.833	0.75	0.782

    #2.0		0.981	0.830	1.2		1		0.9		0.939

    #5.0		1.09	0.923	1.333	1.111	1		1.043

    #10.0	1.045	0.884	1.277	1.064	0.958	1

    # Init coins diameters ratio table, https://en.wikipedia.org/wiki/Israeli_new_shekel
    # coinsDiameters[CoinA][CoinB] = CoinA's diameter / CoinB's diameter.
    coinsDiameters = {}
    coinsDiameters["0.1"] = {}
    coinsDiameters["0.5"] = {}
    coinsDiameters["1.0"] = {}
    coinsDiameters["2.0"] = {}
    coinsDiameters["5.0"] = {}
    coinsDiameters["10.0"] = {}

    coinsDiameters["0.1"]["0.1"] = 1.0
    coinsDiameters["0.1"]["0.5"] = 0.846
    coinsDiameters["0.1"]["1.0"] = 1.22
    coinsDiameters["0.1"]["2.0"] = 1.018
    coinsDiameters["0.1"]["5.0"] = 0.916
    coinsDiameters["0.1"]["10.0"] = 0.956

    coinsDiameters["0.5"]["0.1"] = 1.181
    coinsDiameters["0.5"]["0.5"] = 1.0
    coinsDiameters["0.5"]["1.0"] = 1.444
    coinsDiameters["0.5"]["2.0"] = 1.203
    coinsDiameters["0.5"]["5.0"] = 1.083
    coinsDiameters["0.5"]["10.0"] = 1.130

    coinsDiameters["1.0"]["0.1"] = 0.818
    coinsDiameters["1.0"]["0.5"] = 0.692
    coinsDiameters["1.0"]["1.0"] = 1.0
    coinsDiameters["1.0"]["2.0"] = 0.833
    coinsDiameters["1.0"]["5.0"] = 0.75
    coinsDiameters["1.0"]["10.0"] = 0.782

    coinsDiameters["2.0"]["0.1"] = 0.981
    coinsDiameters["2.0"]["0.5"] = 0.830
    coinsDiameters["2.0"]["1.0"] = 1.2
    coinsDiameters["2.0"]["2.0"] = 1.0
    coinsDiameters["2.0"]["5.0"] = 0.9
    coinsDiameters["2.0"]["10.0"] = 0.939

    coinsDiameters["5.0"]["0.1"] = 1.09
    coinsDiameters["5.0"]["0.5"] = 0.923
    coinsDiameters["5.0"]["1.0"] = 1.333
    coinsDiameters["5.0"]["2.0"] = 1.111
    coinsDiameters["5.0"]["5.0"] = 1.0
    coinsDiameters["5.0"]["10.0"] = 1.043

    coinsDiameters["10.0"]["0.1"] = 1.045
    coinsDiameters["10.0"]["0.5"] = 0.884
    coinsDiameters["10.0"]["1.0"] = 1.277
    coinsDiameters["10.0"]["2.0"] = 1.064
    coinsDiameters["10.0"]["5.0"] = 0.958
    coinsDiameters["10.0"]["10.0"] = 1.0
    
    
    votes = {}
    votes["0.1"] = 0
    votes["0.5"] = 0
    votes["1.0"] = 0
    votes["2.0"] = 0
    votes["5.0"] = 0
    votes["10.0"] = 0

    for i, ratio in enumerate(ratios):
        minDiff = 9999
        v = ""
        hint = hints[i]

        if hint is not None:
            for coinA in coinsDiameters:
                coninB = hint
                if abs(ratio - coinsDiameters[coinA][coninB]) < minDiff:
                    minDiff = abs(ratio - coinsDiameters[coinA][coninB])
                    v = coinA
            votes[v] = votes[v] + 2
            
        else:
            for coinA in coinsDiameters:
                for coninB in coinsDiameters[coinA]:
                    if abs(ratio - coinsDiameters[coinA][coninB]) < minDiff:
                        minDiff = abs(ratio - coinsDiameters[coinA][coninB])
                        v = coinA
                        
            votes[v] = votes[v] + 1
            
    
    print "votes %s" % votes
    
    topVotes = 0
    decision = "UNKNOWN"
    
    for key in votes:
        if votes[key] > topVotes:
            decision = key
            topVotes = votes[key]
            
    return (decision, float(topVotes)/len(ratios))
                    

font = cv2.FONT_HERSHEY_SIMPLEX
# find contours
coins_contours, _ = cv2.findContours(coins_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# make copy of image
coins_and_contours = np.copy(coins)

# find contours of large enough area
min_coin_area = 60
large_contours = [cnt for cnt in coins_contours if cv2.contourArea(cnt) > min_coin_area]

hints = [None] * len(large_contours)

# Ratios
for i in range (len(large_contours)):
    current_contour_area = cv2.contourArea(large_contours[i])
    current_contoure_diameter = cv2.arcLength(large_contours[i], True)

    # contour center
    M = cv2.moments(large_contours[i])
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    
    ratios = []
    print "current_contour_area: %f" % current_contour_area
    for j in range(len(large_contours)):
        ratio = current_contoure_diameter / cv2.arcLength(large_contours[j], True)
        if ratio != 1:
            ratios.append(ratio)
    
    # Match ratio
    print ratios
    label, confidence = MatchRatios(ratios, hints)
    print "confidence %s" % confidence
    if confidence > 0.5:
        print "High confidence value %f for %s" % (confidence, label)
        hints[i] = "%s" % label
        print "hints %s" % hints
        
    print "label: %s" % label
    if label != "UNKNOWN":
        cv2.putText(coins_and_contours, label, (cx, cy), font, 5, (0,255,0), 10)

# draw contours
cv2.drawContours(coins_and_contours, large_contours, -1, (255,0,0), thickness = 6)

coins_and_contours = cv2.cvtColor(coins_and_contours, cv2.COLOR_BGR2RGB)

plt.imshow(coins_and_contours)
```

```python
# create copy of image to draw bounding boxes
bounding_img = np.copy(coins)

# for each contour find bounding box and draw rectangle
for contour in large_contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(bounding_img, (x, y), (x + w, y + h), (0, 255, 0), 10)

bounding_img = cv2.cvtColor(bounding_img, cv2.COLOR_BGR2RGB)
plt.imshow(bounding_img)
```

```python
# find binary image with edges
coins = cv2.imread('images/coins_horizontal.jpg')

# To grayscale
coins_gray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)

# Blur
coins_preprocessed = cv2.GaussianBlur(coins_gray, (11, 11), 0)

# edge detection
coins_edges = cv2.Canny(coins_preprocessed, threshold1=90, threshold2=110)

# print
plt.imshow(cv2.cvtColor(coins_edges, cv2.COLOR_GRAY2RGB))
```

```python
# find hough circles
circles = cv2.HoughCircles(coins_edges, cv2.cv.CV_HOUGH_GRADIENT, dp=1.5, minDist=100, minRadius=160, maxRadius=300)

# Clone original image
coins_circles = np.copy(coins)

# if circles are detected, draw them
if circles is not None and len(circles) > 0:
    # note: cv2.HoughCircles returns circles nested in an array.
    # the OpenCV documentation does not explain this return value format
    circles = circles[0]
    for (x, y, r) in circles:
        x, y, r = int(x), int(y), int(r)
        cv2.circle(coins_circles, (x, y), r, (255, 255, 0), 8)
    plt.imshow(cv2.cvtColor(coins_circles, cv2.COLOR_BGR2RGB))

print('number of circles detected: %d' % len(circles[0]))
```
